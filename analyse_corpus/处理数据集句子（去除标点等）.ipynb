```python
def SeparatePunc(sent):
    new_sent = ''
    for i in range(len(sent)):
        if sent[i] in string.punctuation:
            if i > 0 and i < len(sent) - 1:
                if sent[i] in ",." and sent[i-1].isdigit() and sent[i+1].isdigit():
                    new_sent += sent[i]
                    continue
                if sent[i] == "%" and sent[i-1].isdigit():
                    new_sent += sent[i]
                    continue
                if sent[i] == "$" and (sent[i-1].isdigit() or sent[i+1].isdigit()):
                    new_sent += sent[i]
                    continue
                if sent[i-1] != ' ':
                    new_sent += ' ' + sent[i]
                elif sent[i+1] != ' ':
                    new_sent += sent[i] + ' '
                else:
                    new_sent += sent[i]
            elif i == 0:
                if sent[i] == "$" and sent[i+1].isdigit():
                    new_sent += sent[i]
                    continue
                if sent[i+1] != ' ':
                    new_sent += sent[i] + ' '
                else:
                    new_sent += sent[i]
            else:
                if sent[i] == "%" and sent[i-1].isdigit():
                    new_sent += sent[i]
                    continue
                if sent[i] == "$" and sent[i-1].isdigit():
                    new_sent += sent[i]
                    continue
                if sent[i-1] != ' ':
                    new_sent += ' ' + sent[i]
                else:
                    new_sent += sent[i]
        else:
            new_sent += sent[i]
    return new_sent.strip().lower()


def DeletePuncBlank(sent):
    new_sent = ''
    for i in range(len(sent)):
        if not (sent[i] in string.punctuation or sent[i] == ' '):
            new_sent += sent[i]
    return new_sent
```

```python
def split_dataset_bind_es(corpus, rate):
    # 按照xd的意思根据将专家句子绑在一起; rate:[8,1,1] : train:dev:test
    
    DeletePuncSent = []
    sentences = {i:[] for i in range(1,921)}
    repeats = []
    taskId = [[] for i in range(10)]
    for ci in corpus:
        expertSentence = DeletePuncBlank(ci['expertSentence'])
        if expertSentence not in DeletePuncSent:
            DeletePuncSent.append(expertSentence)
            sentences[ci['expertSentenceId']].append((ci['taskId'], ci['expertSentenceId'], 1, SeparatePunc(ci['expertSentence'])))
            taskId[ci['taskId']-1].append(ci['expertSentenceId'])
        for ri in ci['rewriteSentences']:
            rewriteSentence = DeletePuncBlank(ri)
            if rewriteSentence not in DeletePuncSent:
                DeletePuncSent.append(rewriteSentence)
                sentences[ci['expertSentenceId']].append((ci['taskId'], ci['expertSentenceId'], 0, SeparatePunc(ri)))
                # taskId[ci['taskId']-1].append(len(sentences)-1)
            else: 
                repeats.append(ci['taskId'])
                
    sentences_train = []
    sentences_dev = []
    sentences_test = []
    
    rate1 = (rate[1]+rate[2])/sum(rate)
    rate2 = rate[2]/sum(rate)
    
    for ti in taskId:
        task_len = len(ti)
        dev_test = set(random.sample(ti,round(task_len * rate1)))
        train = set(ti) - dev_test
        dev = set(random.sample(dev_test,round(task_len * rate2)))
        test = dev_test - dev
        for ti in train:
            sentences_train.extend(sentences[ti])
        for di in dev:
            sentences_dev.extend(sentences[di])
        for ti in test:
            sentences_test.extend(sentences[ti])
            
    return sentences_train, sentences_dev, sentences_test
```

```python
sentences_train, sentences_dev, sentences_test = split_dataset_bind_es(corpus, [8,1,1])
len(sentences_train),len(sentences_dev),len(sentences_test)
```
